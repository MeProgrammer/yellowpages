```python
url = "https://www.yellowpages.com.eg/en/condensed-search/dentist?"
html = requests.get(url, headers={"User-Agent": "Mozilla/5.0 (Linux; Android 6.0; HTC One M9 Build/MRA58K) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.98 Mobile Safari/537.3"})
html
```

# now we requested the page 
# and here comes the part of parsing of the html to extract the data

```python
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
dents = soup.find_all('div', attrs={'class': ['searchResultsDiv', 'content-widget']})
```

```python
def get_name_address(elem):
    return [elem.find('a', attrs={'class': ['companyName', 'companyTitle']}).text, elem.find('a', attrs={'class': 'company_address'}).text]
```

```python
get_name_address(dents[0])
```

```python
import pandas as pd
```

```python
ls = []
for dent in dents:
    ls.append(get_name_address(dent))
df = pd.DataFrame(ls, columns= ['Name', 'Address'])
```
